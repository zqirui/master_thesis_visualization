```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mp
from matplotlib import rc
```

```python
eps = 1e-4
cov_x_y = 1.5
var_x_y = 1.74
p = np.random.multivariate_normal(mean=[1.5,1.5], 
                                cov=[[var_x_y,cov_x_y+eps],[cov_x_y+eps,var_x_y]],
                                size=5000)

q = np.random.multivariate_normal(mean=[-1.5,1.5], 
                                cov=[[var_x_y,-cov_x_y+eps],[-cov_x_y+eps,var_x_y]],
                                size=5000)


```

```python
# set seaborn style
sns.set_style("white")

# Basic 2D density plot
fig, ax = plt.subplots(1,2, figsize=(15,6))
# kde plot P
sns.kdeplot(x=p[:,0], y=p[:,1], color="red", ax=ax[0], bw_adjust=1.25)
# kde plot Q
sns.kdeplot(x=q[:,0], y=q[:,1], color="blue", ax=ax[0], bw_adjust=1.25)

sns.kdeplot(x=p[:,0], y=p[:,1], color="red", ax=ax[1], bw_adjust=1.25, levels=2, fill=True, alpha=0.45)
sns.kdeplot(x=q[:,0], y=q[:,1], color="blue", ax=ax[1], bw_adjust=1.25, levels=2, fill=True, alpha=0.45)
ax[1].text(x=-0.25, y=0, s=r'$\mu$')
ax[1].text(x=-2.5, y=2.5, s=r'$v_Q$')
ax[1].text(x=2.5, y=2.5, s=r'$v_P$')


handles = [mp.Patch(color="red", label="P"),
           mp.Patch(color="blue", label="Q")]

for axes in ax:
    axes.set(xticklabels=[])
    axes.set(yticklabels=[])
    axes.legend(handles=handles)



ax[0].title.set_text("Example densities of P and Q")
ax[1].title.set_text(r"Graphical interpretation of $\mu$,$v_P$ and $v_Q$")
plt.savefig("PRD_1.svg", dpi=300)
plt.show()
```
