```python
import numpy as np
from scipy.stats.qmc import Sobol
import matplotlib.pyplot as plt

sobol_sampler = Sobol(d=2, scramble=False)
samples = sobol_sampler.random_base2(m=10) # 2^m in [0,1]
print(samples.shape)

uniform_samples = np.random.uniform(0,1.0, size=(1024,2))
print(uniform_samples.shape)

fig, axes = plt.subplots(1,2, figsize=(9,4))
axes[0].scatter(x=uniform_samples[:,0], y=uniform_samples[:,1], s = 5)
axes[0].set_title("Uniform Sampling")
axes[1].scatter(x=samples[:,0], y=samples[:,1], s = 5)
axes[1].set_title("Sobol Sampling")
plt.savefig("Sobol.svg", dpi=300)
plt.show()
```
